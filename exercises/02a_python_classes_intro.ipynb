## Section 0: What *is* a class? What *is* an object?

**You will learn:** the difference between a class and an object.

A **class** is a **blueprint**: it describes what every object built from it will have (its data) and what it can do (its behavior). From one blueprint you can build many real objects.

In OOP:
- a **class** is the blueprint (e.g. `Animal`),
- an **object** (also called an **instance**) is a concrete thing built from it (e.g. Simba, a 5 year old lion).

| Real world        | OOP term  | Example                          |
|-------------------|-----------|----------------------------------|
| Blueprint         | class     | `Animal`                         |
| One real animal   | object    | `simba = Animal("Simba", 5)`     |
| The animal's age  | attribute | `simba.age`                      |
| What it can do    | method    | `simba.eat()`                    |

Below is a tiny `Animal` blueprint and two objects built from it.

````python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

simba = Animal("Simba", 5)
nala = Animal("Nala", 4)
print(simba.name, simba.age)
print(nala.name, nala.age)
````

## Section 1: Attributes and `__init__`

**You will learn:** how to define a class with `__init__`, attributes and `self`.

The special method **`__init__`** (the *constructor*) runs **automatically** every time you create a new object. Its job is to set up the object's starting **attributes** (its data).

Two things to notice:
- **`self`** is the first parameter of every method. It refers to *the particular object being created or used*. `self.name = name` means store the given name on this object.
- The attributes `name`, `age`, `hunger`, `energy` together form the object's **state**, the data that describes it right now.

Now you build the real `Animal` constructor. From here on we keep growing this same class, we never rewrite it from scratch.

### Your turn: write the constructor

Complete `__init__` so that a new `Animal` stores `name` and `age`, and starts with `hunger` set to `5` and `energy` set to `5`. These four attributes are the animal's starting state.

````python
# 🎯🎯🎯 TODO #2 🎯🎯🎯: in __init__ store name and age, and set hunger and energy to 5
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.hunger = 5
        self.energy = 5
````

## Section 2: Methods, giving animals behavior

**You will learn:** how methods read and change an object's state through `self`, and that a method is just a function attached to a class.

Attributes are what an object *knows*. **Methods** are what an object *can do*: they are functions that live on the class and act on `self`.

You already built `Animal` with its `__init__`. We do **not** rewrite it. Instead we **add** behavior to the same class: define a function and attach it with `Animal.eat = eat`. Here `eat` and `sleep` change the state, and `make_sound` returns a short string. Then you add one more method the same way.

````python
def eat(self):
    self.hunger = max(0, self.hunger - 3)
    self.energy = max(0, self.energy - 1)

def sleep(self):
    self.energy = 10
    self.hunger = min(10, self.hunger + 2)

def make_sound(self):
    return f"{self.name}: ..."

Animal.eat = eat
Animal.sleep = sleep
Animal.make_sound = make_sound

simba = Animal("Simba", 5)
print(simba.make_sound())
simba.eat()
print("After eating, hunger:", simba.hunger)
simba.sleep()
print("After sleeping, energy:", simba.energy)
````

## Section 3: Inheritance, `Dog`, `Cat`, `Zebra` (exercise optional)

**You will learn:** how a child class reuses a parent class with inheritance and `super()`.

Every cat, dog and zebra has a name, an age, gets hungry and tired, exactly like our `Animal`. Rewriting all of that for each one would be repetitive and error prone.

**Inheritance** lets a **child class** reuse everything from a **parent class**. We write `class Dog(Animal):` to say a Dog **is an** Animal, and gets all of Animal's attributes and methods for free. A child can both **reuse** the parent and **add** behavior of its own.

Each child here defines its own `__init__` that calls **`super().__init__(...)`** to reuse the shared setup, then adds a bit of its own state: the dog counts how many times it has played, the cat counts its naps, the zebra stores its stripes. Each child also adds a method that really does something, not just returns a sentence: `Dog.play` makes the dog tired and hungry, `Cat.nap` restores energy, and `Zebra.has_more_stripes_than` compares two zebras and returns `True` or `False`. Then you add a new species.

````python
class Dog(Animal):
    def __init__(self, name, age):
        super().__init__(name, age)
        self.times_played = 0

    def play(self):
        self.energy = max(0, self.energy - 2)
        self.hunger = min(10, self.hunger + 2)
        self.times_played += 1

class Cat(Animal):
    def __init__(self, name, age):
        super().__init__(name, age)
        self.naps_taken = 0

    def nap(self):
        self.energy = min(10, self.energy + 4)
        self.naps_taken += 1

class Zebra(Animal):
    def __init__(self, name, age, stripes):
        super().__init__(name, age)
        self.stripes = stripes

    def has_more_stripes_than(self, other):
        return self.stripes > other.stripes

rex = Dog("Rex", 3)
print("Rex energy", rex.energy, "hunger", rex.hunger, "times played", rex.times_played)
rex.play()
print("After playing, energy", rex.energy, "hunger", rex.hunger, "times played", rex.times_played)

whiskers = Cat("Whiskers", 2)
whiskers.energy = 4
whiskers.nap()
print("After a nap, Whiskers energy", whiskers.energy, "naps taken", whiskers.naps_taken)

marty = Zebra("Marty", 6, stripes=42)
zara = Zebra("Zara", 4, stripes=30)
print("Marty has more stripes than Zara?", marty.has_more_stripes_than(zara))
````

## Section 4: Method overriding (Optional)

**You will learn:** how a child class replaces an inherited method, and what polymorphism means.

Right now every animal returns the sound `"<name>: ..."`. But a dog should bark and a cat should meow. A child class can **override** an inherited method, that is give it its **own version** with the **same name**.

The example shows two ways to override. We give the existing `Dog`, `Cat` and `Zebra` their own sound by attaching it, and we define a new class `Cow` that overrides `make_sound` right in its body. Then we loop over a list of different animals and call the **same** method name on each. Each object does its own thing. This idea is called **polymorphism**. Overriding works for any method, not only `make_sound`, and that is what you will practise next.

````python
def woof(self):
    return f"{self.name}: Woof!"

def meow(self):
    return f"{self.name}: Meow!"

def neigh(self):
    return f"{self.name}: Neigh!"

Dog.make_sound = woof
Cat.make_sound = meow
Zebra.make_sound = neigh

class Cow(Animal):
    def make_sound(self):
        return f"{self.name}: Moo!"

for a in [Dog("Rex", 3), Cat("Whiskers", 2), Zebra("Marty", 6, stripes=42), Cow("Bessie", 4)]:
    print(a.make_sound())
````

## Section 5: People, and the `Zoo` that holds everything

**You will learn:** a second class hierarchy and composition, where one object contains others.

A zoo is not only animals. We add an independent hierarchy of **people**: a `Visitor` who watches animals and a `Zookeeper` who feeds them. A `Person` is *not* an `Animal`, so it has its own parent class. Keeping these jobs apart is called **separation of responsibilities**.

````python
class Person:
    def __init__(self, name):
        self.name = name

class Visitor(Person):
    def look_at(self, animal):
        return f"{self.name} watches {animal.name} ({animal.make_sound()})"

class Zookeeper(Person):
    def feed(self, animal):
        animal.eat()
        return f"{self.name} fed {animal.name}. Hunger is now {animal.hunger}."

joe = Zookeeper("Joe")
amy = Visitor("Amy")
rex = Dog("Rex", 3)
print(joe.feed(rex))
print(amy.look_at(rex))
````

### First: an object can be an attribute of another object

So far every attribute has been a **simple value**, a string like `name` or a number like `age` and `hunger`. But an attribute can also be **another object**, and that is the one new idea behind the `Zoo` you are about to build. Let's see the smallest version first: a single object stored inside another, before we hold many of them in a list.

In the example below a `Lion` stores an `Enclosure` **object** in `self.enclosure`. Because that attribute is itself an object, we reach *through* it with a second dot: `leo.enclosure.climate`. So `leo` is an object, `leo.enclosure` is another object living inside it, and `leo.enclosure.climate` reaches the data inside that second object. The `Zoo` is exactly this idea, but holding a **list** of objects instead of a single one.

````python
# A tiny preview: an object stored *inside* another object

class Enclosure:
    def __init__(self, climate):
        self.climate = climate

class Lion:
    def __init__(self, name, enclosure):
        self.name = name
        self.enclosure = enclosure   # this attribute is itself an OBJECT

savanna = Enclosure("hot and dry")   # 1) build the inner object
leo = Lion("Leo", savanna)           # 2) pass it in, stored as an attribute

print(leo.name)               # Leo
print(leo.enclosure.climate)  # hot and dry  -> object inside object, two dots
````

> a zoo *has* animals leads to composition
